```python
#xlsum

import re
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

WHITESPACE_HANDLER = lambda k: re.sub('\s+', ' ', re.sub('\n+', ' ', k.strip()))

article_text = '''คือ เทียบกับรูปในเว็บtaobao ต่างกันนิดหน่อยเองแค่ปลายฐาน นอกนั้นเหมือนเลย แล้วจะสิ่งประดิษฐ์ยังไง #กรมทรัพย์สินทางปัญญา #ปังชา เข้าใจว่ามันเป็นฟีลน้ำแข็งใสราดเฮบลูบอย ใครๆก็ทำกัน ไม่ใช่เหรอ การตลาดน่าจะขึ้นอยู่กับน้ำที่ราดว่าเจ้าไหนอร่อยกว่าป่าว เกิดมาเพิ่งเคยเจอ #ปังชา อยากรู้ว่าเจ้าของแบรนด์เขาจะยังภูมิใจกับแบรนด์ที่ตัวเองปั้นมาอยู่มั้ย โดนด่าเยอะมาก สงสารในมุมคนที่ปั้นแบรนด์มา แต่สงสารร้านที่โดนฟ้องมากกว่า #ปังชา #ลูกไก่ทอง ไปรายงานเพจมัน #ปังชา เพื่อนกุชอบกินมากจ้าาาาา ตอนนี้ยังจะชอบกินอยุ่ไหมนิ 55555555 #ปังชา มากทม ทีไรชวนไปตลอด แต่ไม่เคยไปเลยนะ มันแพงเกิน ตังไม่มี ระวังเจอเคสแบบ โรงแรมนะเดือดร้อนทั้งองค์กรเลย อย่ามาแหย่หนวดชาวเน็ตจะดีกว่า #ปังชา จากเคสของ #ปังชา ไอเดียบรรเจิดเลย(เรื่องสมมุติ)กูจะตั้งชื่อร้านปังเย็น(เพราะชอบกินปังเย็นใน7-11) เมนูปังเย็นทั้งหลายกูจะไปจดลิขสิทธิ์ให้หมด แล้วฟ้อง all café บ้างดีไหม หากินง่ายดีเปิดร้านไว้บังหน้า หรือเจ้าสั่วจะฟ้องกูก่อน ร้านลูกไก่ ส่วนตัวไม่ชอบอ่ะ portion ใหญ่ไป ชาไม่ทั่วน้ำแข็ง หวานไม่ถึง งงนะว่าทำไมได้มิชลิน ลองครั้งเดียวก้เกินพอ #ปังชา เหตุผลเหี้ยอ่ะ ที่เรียกขนาดนั้นแค่อยากให้เขาตกใจจะได้หยุด" คือเหี้ยกว่าเดิม คนสันดานเหี้ยเท่านั้นถึงคิดได้อ่ะ เขาเป็นผู้ถูกกระทำด้วยซ้ำ แล้วเมื่อวานที่บอกสื่อสารผิดคือตอแหลชัดเจนเนาะ เก่งแต่กับคนตัวเล็กๆ #ลูกไก่ทอง #ปังชา" อ่ะ สรุปแล้วจบนะ #ปังชา'''

model_name = "csebuetnlp/mT5_multilingual_XLSum"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

input_ids = tokenizer(
    article_text,
    return_tensors="pt",
    padding="max_length",
    truncation=True,
    max_length=512
)["input_ids"]

output_ids = model.generate(
    input_ids=input_ids,
    max_length=84,
    no_repeat_ngram_size=2,
    num_beams=4
)[0]

summary = tokenizer.decode(
    output_ids,
    skip_special_tokens=True,
    clean_up_tokenization_spaces=False
)

print(summary)
```
